```python
car_sales["Price"] = car_sales["Price"].str.replace('[\$\,\.]', '')
car_sales
```

```python
car_sales["Price"] = car_sales["Price"].str[:-2]
car_sales
```

```python
car_sales["Sales Date"] = pd.date_range('1/1/2023', periods =len(car_sales))
car_sales
```

```python
car_sales["Total Sale"] = car_sales['Price'].astype(int).cumsum()
car_sales
```

```python
car_sales.plot(x='Sales Date', y="Total Sale", xlabel="Total sale", ylabel = "Sale Date", title='ars Sales in 2023');
# car_sales.set(xlabel="Total sale", ylabel = "Sale Date", title='cars Sales in 2023')
```

```python
heart_csv = pd.read_csv("heart-disease.csv")
heart_csv.head()
```

```python
heart_csv["age"].plot.hist(bins=10);
```

```python
heart_csv.head()
```

```python
heart_csv.plot.hist(figsize=(10, 30), subplots=True);
```

```python
over_50 = heart_csv[heart_csv["age"] >50 ]
over_50.head()
```

```python
over_50.plot(kind = 'scatter',
            x='age',
            y='chol',
            c='target');
```

```python
fig, ax = plt.subplots(figsize=(10,6))
over_50.plot(kind = 'scatter',
            x='age',
            y='chol',
            c='target',
            ax =ax);
ax.set_xlim([45,100]);
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
scatter = ax.scatter(x=over_50['age'],
                    y=over_50['chol'],
                    c=over_50['target']);
ax.set(title="heart disease and cholestrol levels",
      xlabel='Age',
      ylabel='cholestrol');
ax.legend(*scatter.legend_elements(), title='target');
ax.axhline(over_50['chol'].mean(),linestyle='--');
```

```python
x = np.linspace(0, 2, 100)  # Sample data.

# Note that even in the OO-style, we use `.pyplot.figure` to create the Figure.
fig, ax = plt.subplots(figsize=(10, 10), layout='constrained')
ax.plot(x, x, label='linear')  # Plot some data on the axes.
ax.plot(x, x**2, label='quadratic')  # Plot more data on the axes...
ax.plot(x, x**3, label='cubic')  # ... and some more.
ax.set_xlabel('x label')  # Add an x-label to the axes.
ax.set_ylabel('y label')  # Add a y-label to the axes.
ax.set_title("Simple Plot")  # Add a title to the axes.
ax.legend();  # Add a legend.
```
